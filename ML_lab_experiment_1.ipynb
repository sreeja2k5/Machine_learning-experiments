```python


import statistics
```

```python
exam_scores = [78, 85, 90, 88, 76, 95, 89, 92, 85, 87]
```

```python
mean_value = statistics.mean(exam_scores)
median_value = statistics.median(exam_scores)
mode_value = statistics.mode(exam_scores)
```

```python
variance_value = statistics.variance(exam_scores)
std_dev_value = statistics.stdev(exam_scores)
```

```python
print(f"Mean (Average Score): {mean_value}")
print(f"Median (Middle Score): {median_value}")
print(f"Mode (Most Frequent Score): {mode_value}")
print(f"Variance (Score Spread): {variance_value}")
print(f"Standard Deviation (Score Deviation): {std_dev_value}")
```
